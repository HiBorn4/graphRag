```python
import pandas as pd
```

```python
# Path to your Excel file
original_excel = "inventories_compare_new.xlsx"


# Load both sheets into separate DataFrames
# df_fy23 = pd.read_excel(excel_path, sheet_name="Schedules Mar F-23")
# df_fy24 = pd.read_excel(excel_path, sheet_name="Schedules Mar F-24")
original_excel_df_fy24_new = pd.read_excel(original_excel, sheet_name="base_tb_inventory_24")
original_excel_df_fy23_new = pd.read_excel(original_excel, sheet_name="base_tb_inventory_23")
```

```python
original_excel_df_fy23_new.head()
```

```python
def clean_and_filter_df(df):
    # Remove dots from 'Text for B/S P&L item' column if it exists
    if "Text for B/S P&L item" in df.columns:
        df["Text for B/S P&L item"] = df["Text for B/S P&L item"].str.replace(".", "", regex=False)

    # Columns to retain
    keep_cols = [
        '001-AS', '133-FES', '008-SWARAJ', '002-HO', 'Powerol', 'CE', 'Agri', '004-MDS',
        'Synergy', '007-SBU', 'TW', '134-TPDS', '135-INTNL OPER', '138-CONT SOUR',
        '140-NON MRV', '141-SOURCING', '139-CORP'
    ]
    
    # Filter only those columns that exist in the dataframe
    keep_cols_filtered = [col for col in keep_cols if col in df.columns]

    return df[keep_cols_filtered]
```

```python
inventories_fy23_values_df = clean_and_filter_df(original_excel_df_fy23_new)
inventories_fy24_values_df = clean_and_filter_df(original_excel_df_fy24_new)
```

```python
new_excel = "inventories_compare.xlsx"
new_df_fy24_new = pd.read_excel(new_excel, sheet_name="base_tb_inventory_24")
new_df_fy23_new = pd.read_excel(new_excel, sheet_name="base_tb_inventory_23")
```

```python
new_df_fy24_new.head()
```

```python
def trim_and_rename(df):
    df = df.iloc[:, :4]
    df.columns = ['L1', 'L2', 'L3', 'L4']
    return df
```

```python
new_df_fy23_new = trim_and_rename(new_df_fy23_new)
new_df_fy24_new = trim_and_rename(new_df_fy24_new)
```

```python
merged_df_fy23 = pd.concat([new_df_fy23_new, inventories_fy23_values_df], axis=1)
merged_df_fy24 = pd.concat([new_df_fy24_new, inventories_fy24_values_df], axis=1)
```
